## DATA PREPROCESSING

### Tasks for data cleaning
- sample is reduced to 10000
- remove html tags
- remove special characters
- convert to lower case
- remove stopwords
- stemming

```python
df=df.sample(10000)
```

```python
df.shape
```

```python
df.info()
```

```python
df['sentiment'].replace({'positive':1,'negative':0}, inplace=True)
```

```python
df.head()
```

```python
import re
```

```python
# function to clean html tags
def clean_html(text):
    clean=re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review']=df['review'].apply(clean_html)
```

```python
# function to convert to lower case
def convert_lower(text):
    return text.lower()
```

```python
df['review']=df['review'].apply(convert_lower)
```

```python
# function to remove special characters
def remove_special(text):
    x= ''
    for i in text:
        if i.isalnum():
            x = x+i
        else:
            x = x+' '
    return x
```

```python
df['review']=df['review'].apply(remove_special)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
# function for removing stopwords
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    
    return y
```

```python
df['review']=df['review'].apply(remove_stopwords)
df
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# function for stemming
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    
    return z
```

```python
df['review']=df['review'].apply(stem_words)
df
```

```python
# join
def join(list_input):
    return " ".join(list_input)
```

```python
df['review']=df['review'].apply(join)
df
```

```python
x = df['review']
y = df['sentiment']
```

```python
x.shape,y.shape
```

### TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=50000)
tfidf_X = tfidf.fit_transform(x).toarray()
tfidf_X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(tfidf_X, y, test_size=0.20)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model1 = MultinomialNB()
model1.fit(X_train,y_train)
```

```python
print("MultinomialNB ",accuracy_score(y_test,model1.predict(X_test)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cm1 = confusion_matrix(y_test,model1.predict(X_test))
```

```python
cm_matrix1 = pd.DataFrame(data=cm1, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                    index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix1, annot=True, fmt='d', cmap='YlGnBu')
```

```python
print(classification_report(y_test,model1.predict(X_test)))
```
